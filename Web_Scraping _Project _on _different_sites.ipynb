<h2 align=center style=color:red>  Scraping a  Population Table from Worldometers website<h2/>

```python
url = "https://www.worldometers.info/world-population/" # This site is Worldometers which tell the real time population of world
```

```python
page = requests.get(url)
```

```python
page        #if it is 200 then it means we can scrap data from this site.
```

```python
html = BeautifulSoup(page.text, "lxml")
```

```python
table = html.find("table","table table-striped table-bordered table-hover table-condensed table-list")
```

```python
table
```

```python
t_headers = table.find_all("th")
```

```python
t_headers
```

```python
headers = []
for i in t_headers:
    title = i.text
    headers.append(title)
```

```python
headers
```

```python
df = pd.DataFrame(columns = headers)
```

```python
df
```

#### Using nested method to scrap the rows of a table while leaving header

```python
table.find_all("tr")[1:]
```

```python
for j in table.find_all("tr")[1:]:
    row_data = j.find_all("td")
    row = [tr.text for tr in row_data]
    length = len(df)
    df.loc[length] = row
     
```

```python
df   
```

#### Exporting this table into csv or excel

```python
df.to_csv('table_scraped.csv')
```

```python
df.to_excel("table_scrapped.xlsx")
```

```python
df.to_excel("table_scraped.xlsx", index = False)
```

<h2 align=center style=color:green> Scraping a 2022  NFL League Standings Table <h2/>

```python
 url = "https://www.nfl.com/standings/league/2022/REG"
```

```python
page1 = requests.get(url)
```

```python
nfl = BeautifulSoup(page1.text , "lxml")
```

```python
table1 = nfl.find("table",{"summary":"Standings - Detailed View"})
```

```python
header = nfl.find_all("th")
```

```python
header_list = []
for i in header:
    list = i.text.strip()
    header_list.append(list)
```

```python
header_list
```

```python
df1 = pd.DataFrame(columns = header_list)
```

```python
df1
```

```python
 tr1 = table1.find_all('tr')[1:]
```

```python
for j in tr1:
    first_td = j.find_all("td")[0].find("div",class_ ="d3-o-club-fullname").text.strip()
    row_data = j.find_all("td")[1:]
    data = [td.text.strip() for td in row_data]
    data.insert(0,first_td)
    length1 = len(df1)
    df1.loc[length1] = data
```

```python
df1.to_csv("table1_scraped.csv")
```

```python
### Imorting libraries
import csv
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
# Creating an instance of webdriver for google chrome
driver = webdriver.Chrome()
```

```python
# Using webdriver we'll now open the dermnetnz website in chrome
url =  'https://dermnetnz.org/image-library'
# We;ll use the get method of driver and pass in the URL
driver.get(url)
```

```python
print(url)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
# Total no. of  wanted data present in this page
results=soup.find_all('a',{'class':"imageList__group__item"})
len(results)
```

```python
# picking the 1st card from the complete list of cards
item = results[0]
```

###  Extracting data for single diseases

```python
# Extracting the name of Diseases from the 1st card
name = item.find('div',{'class':"imageList__group__item__copy"}).text.strip()
name
```

```python
# Extracting the urls of Diseases from the 1st card
urls='https://dermnetnz.org' + soup.find('a' ,{'class':"imageList__group__item"},href=True)['href']
urls
```

###  Extracting all data points of name and urls

```python
# Define a function for both name and urls for single data
def extract_name_url(item):
    name = item.find('div',{'class':"imageList__group__item__copy"}).text.strip()
    urls='https://dermnetnz.org' + soup.find('a' ,{'class':"imageList__group__item"},href=True)['href']
    file=(name,urls)
    return file
```

```python
# creating a list of all name and urls
lists = []
results = soup.find_all('a',{'class':"imageList__group__item"})
for item in results:
    lists.append(extract_name_url(item))
```

```python
# creating a pandas dataframe
pd.DataFrame(lists,columns=['name','urls'])
```

```python
# save file into csv
pd.DataFrame(lists,columns=['name','urls']).to_csv('Diseases_name and urls')
```

###  Extracting all urls of image

```python
# finding all img html tag
image_tags = soup.find_all('img')
```

```python
# Creating a list of urls of images
links = []
for image_tag in image_tags:
    try:
        links.append(image_tag['src'])
    except:
        pass
```

<h2 align=center style=color:blue> Scraping Airbnb Site and dealing with multiple pages website<h2/>

```python
url1 =   "https://www.airbnb.co.in/s/Goa--India/homes?tab_id=home_tab&refinement_paths%5B%5D=%2Fhomes&flexible_trip_lengths%5B%5D=one_week&price_filter_input_type=0&price_filter_num_nights=1&query=Goa%2C%20India&place_id=ChIJQbc2YxC6vzsRkkDzYv-H-Oo&date_picker_type=calendar&checkin=2023-01-20&checkout=2023-01-21&source=structured_search_input_header&search_type=autocomplete_click&federated_search_session_id=56ff82a7-a82f-409d-b469-7b0860fc220e&pagination_search=true&cursor=eyJzZWN0aW9uX29mZnNldCI6MCwiaXRlbXNfb2Zmc2V0IjowLCJ2ZXJzaW9uIjoxfQ%3D%3D"
```

```python
page5 = requests.get (url1)
```

```python
page5
```

```python
airbnb = BeautifulSoup(page5.text , "lxml")
```

```python
df6 = pd.DataFrame({"Link":[""],"Title" :[""], "Detail":[""],"Rating" :[""],"Price":[""],"Address":[""]})    
```

#### Forming the loop for multiple pages

```python
counter = 0

while counter <15:
    

    postings = airbnb.find_all("div", class_ = "c4mnd7m dir dir-ltr" )
    for post in postings:
        try:


            link = post.find("a", class_ ="bn2bl2p dir dir-ltr").get("href")
            link_full = "https://www.airbnb.co.in" + link
            title = post.find("div", class_ = "t1jojoys dir dir-ltr" ).text
            price = post.find("span" , class_ ="a8jt5op dir dir-ltr").text
            rating = post.find("span", class_ = "r1dxllyb dir dir-ltr").text
            details = post.find("span" , class_ = "dir dir-ltr").text
            address = post.find("span", class_ = "t6mzqp7 dir dir-ltr").text
            df6 = df6.append({"Link":link_full,"Title" :title, "Detail":details,"Rating" :rating,"Price":price,"Address":address},ignore_index = True)
        except:
            pass






    next_page = airbnb.find("a",{"aria-label" :"Next"}).get("href")
    next_page_full = "https://www.airbnb.co.in" + next_page

    
    page5 = requests.get(next_page_full)
    airbnb = BeautifulSoup(page5.text, "lxml")
    counter += 1
```

###  Saving file into excel and csv

```python
df6.to_csv("rooms_scraped.csv")
```

```python
df6.to_excel("rooms_scraped.xlsx")
```

<h2 align=center style=color:red> Scraping Carpages site and dealing with multiple pages <h2/>

```python
url10  = "https://www.carpages.ca/used-cars/search/?fueltype_id%5B0%5D=3&fueltype_id%5B1%5D=7"
```

```python
page10 = requests.get(url10)
```

```python
page10
```

```python
carpages = BeautifulSoup(page10.text,"lxml")
```

```python
postings10 = carpages.find_all("div", class_ = "media soft push-none rule")
```

```python
df10 = pd.DataFrame({"Link":[""],"Name":[""],"Price":[""],"KM_Run":[""],"Color":[""]})
```

#### Forming the loop for multiple pages

```python
counter = 0  
    
while counter <15:
    postings10 = carpages.find_all("div", class_ = "media soft push-none rule")
    
    for post in postings10:
            link = post.find("a",{"class" :"media__img media__img--thumb"}).get("href")
            link_full = "https://www.carpages.ca" + link
            name = post.find("h4", class_ = "hN").text.strip()
            price = post.find("strong", class_ = "delta").text
            km_run = post.find_all("div", {"class" :"grey l-column l-column--small-6 l-column--medium-4"})[0].text.strip()
            color = post.find_all("div", {"class" :"grey l-column l-column--small-6 l-column--medium-4"})[1].text.strip()
            df10 = df10.append({"Link": link_full ,"Name":name ,"Price":price,"KM_Run":km_run,"Color":color},ignore_index = True)

    next_page = carpages.find("a",{"class":"nextprev"}).get("href")
    next_page_full = "https://www.carpages.ca" + next_page
    page10 = requests.get(next_page_full)
    carpages = BeautifulSoup(page10.text,"lxml")
    counter += 1
```

```python
df10
```

### Saving file into excel and csv

```python
df10.to_csv("cars_scraped.csv")
```

```python
df10.to_excel("cars.scraped.xlsx")
```
